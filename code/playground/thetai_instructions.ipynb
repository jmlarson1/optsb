# Basic Description and Goal

This example will showcase the running of TRACK for the transport beam line.
There will be output files for comparisions to ensure TRACK is running properly.
The `test_track.py` file runs TRACK and is where the data set is selected (the i).

# Running `test_track.py`

All the relative directory links assumes running from the `/code/playground` directory, but some hard links may still exist.

```bash
cd /optsb/code/playground
```

If wine and libraries are installs, can just exec `test_track.py`:

```bash
python test_track.py
```

Inside `test_track.py` there is a variable to select which `theta_i` data set to execute,

```bash
# CHOOSE WITH THETA_## from 01 to 03
thetai_name="01"
```

This will setup the `run_track.py` to run everything in the following folder:
```bash
folder_location="transport_line/theta_"+thetai_name+"working"
```

Note, that actually what happens is that the `sclinac.dat` file is first copied to `/working/` then executed there.

The `sclinac.dat` file in `/theta_XX/` is the file to modify to change the quad settings for each of the 12 quad params.
For example, the first three quads are shown below, where the first is labeled as `1 quad VALUE ...`

```bash
!QTA001 X Y X                                                                   
1  drift    5.0     2.984   2.984                                               
1  quad     -5400.06679 25.00   25.00    2.98   0     100
1  drift    10.0     2.984   2.984
2  quad      4473.06053   25.00   25.00    2.98   0     100
1  drift    10.0     2.984   2.984
3  quad     -2968.62094   25.00   25.00    2.98   0     100
1  drift    5.0     2.984   2.984
```                                          

I have added the ranges for each of the 12 params in the overleaf.

A number of output files are created in `/code/track/transport_line/theta_XX/working/` the main ones of interest are 

`beam.out` gives beam properties for each item listed in sclinac.dat file. 
Last two columns are number of particles lost and left to calculate transmission.

`coord.out` gives particle values for particles that made it to the target. File only includes 
the particles that made it so number / length of file is the transmission as 10,000 particles are 
run for each simulation (10k).

Along with a number of output files created in `/code/track/transport_line/theta_XX/working/` 
a figure is generated as `profile00001.png`
