# Basic Description and Goal

This example will showcase the running of TRACK for the sps beam line. 
It includes the modification of the quadrupole parameters as well (which are the `actions` in the gym env language).
There will be output files for comparisions to ensure TRACK is running properly.
The `test_track.py` file runs TRACK and interactions with input / output files some.s
The `data_generator.py` file is a rough start to interacting with TRACK via the gym-env framework.

# Required Installs

All codes (except TRACK) are in python. Hence, various standard python packages / libraries are required.

Also, I added the `gym_optsb` locally via ``pip3 install -e gym-optsb`` inside the `optsb/code/gym_envs` directory

To run TRACK, a working version of `wine` or `wine64` is needed. I (CRH) have a basic install of `wine` on my MAC via `homebrew`
I have also tested wine64 on an Ubuntu install
TRACK exe and all .dll files will be pulled via git.

# Running `test_track.py`

All the relative directory links assumes running from the /code/playground directory

```bash
cd /optsb/code/playground
```

If wine and libraries are installs, can just exec `test_track.py`:

```bash
python test_track.py
```
The code will setup a new directory `/code/track/sps_line/testing/` and cp all TRACK files that are needed for input.
It will then run one instance of TRACK setting all magnets to =0.

## Running with Random Magnet Settings

To run TRACK with random magnet settings, simply uncomment line 33 in `test_track.py` and comment out 34

```bash
    quad_vals = rs.get_quad_vals() #[1100,-1900,1200]
    #quad_vals = [0,0,0]
```

Now three random values will be used for the three quads.
